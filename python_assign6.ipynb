```python
# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

np.random.seed(0)

n_samples = 100
x1 = np.random.normal(0, 1, n_samples) 
x2 = 2 * x1 + np.random.normal(0, 0.5, n_samples)  
x3 = 0.5 * x1 + 0.5 * x2 + np.random.normal(0, 0.3, n_samples) 

df = pd.DataFrame({'Feature 1': x1, 'Feature 2': x2, 'Feature 3': x3})

corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True, vmin=-1, vmax=1)

plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# 4. Using the given x and y data, generate a 3D surface plot to visualize the function 

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

import plotly.graph_objects as go

fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

fig.update_layout(
    title='3D Surface Plot of f(x, y) = sin(sqrt(x^2 + y^2))',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    ),
    autosize=True
)

fig.show()
```

```python
# Bokeh Assignment
# 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x

from bokeh.plotting import figure, show
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)

p.line(x, y, legend_label="sin(x)", line_width=2, line_color="blue")

show(p)
```

```python
# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

from bokeh.plotting import figure, show
import numpy as np
import pandas as pd

np.random.seed(42)
n = 100
x = np.random.rand(n) * 10  
y = np.random.rand(n) * 10  
sizes = np.random.randint(10, 100, n)  
colors = np.random.choice(['red', 'green', 'blue', 'orange'], n)  

df = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

p = figure(title="Scatter Plot with Variable Sizes and Colors", 
           x_axis_label='X', y_axis_label='Y', 
           width=800, height=400)

p.scatter('x', 'y', source=df, size='sizes', color='colors', legend_field='colors', fill_alpha=0.6)

show(p)
```

```python
# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

from bokeh.plotting import figure, show
from bokeh.io import output_file
import pandas as pd

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
df = pd.DataFrame({'fruits': fruits, 'counts': counts})

p = figure(x_range=df['fruits'], title="Fruit Count Bar Chart", 
           x_axis_label='Fruit', y_axis_label='Count', width=800, height=400)

p.vbar(x='fruits', top='counts', source=df, width=0.5, color="skyblue", legend_field="fruits")

show(p)
```

```python
# 4. Create a Bokeh histogram to visualize the distribution of the given data

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
p = figure(title="Histogram of Random Data", 
           x_axis_label='Value', y_axis_label='Frequency', 
           width=800, height=400)

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], 
       fill_color="skyblue", line_color="white", alpha=0.7)

show(p)
```

```python
# 5. Create a Bokeh heatmap using the provided dataset.

from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.models import ColorBar
from bokeh.transform import linear_cmap
from bokeh.layouts import column
from bokeh.palettes import Viridis256
from bokeh.models import LinearColorMapper
import numpy as np
import pandas as pd

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

x_values = xx.flatten()
y_values = yy.flatten()
z_values = data_heatmap.flatten()

df = pd.DataFrame({'x': x_values, 'y': y_values, 'z': z_values})

color_mapper = LinearColorMapper(palette=Viridis256, low=min(z_values), high=max(z_values))

p = figure(title="Heatmap", 
           x_axis_label='X', y_axis_label='Y', 
           width=600, height=600)

p.rect(x='x', y='y', width=0.1, height=0.1, source=df, 
       fill_color={'field': 'z', 'transform': color_mapper}, 
       line_color=None)

color_bar = ColorBar(color_mapper=color_mapper, width=8, location=(0,0))
p.add_layout(color_bar, 'right')

show(p)
```
